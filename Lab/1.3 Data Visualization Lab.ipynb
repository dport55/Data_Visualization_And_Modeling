**Hint:**  Import the Spotify data using `df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')`.

```python
import pandas as pd

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')

df.head()
```

```python
MC_df = df.loc[df['track_name'] == 'Flowers']
MC_df.head()
```

```python

filtered_df = df [(df['bpm'] >= 120) & (df['bpm'] <= 130) & (df['mode'] == 'Minor') & (df['key'] == 'A') ]

filtered_df.head()
```

```python

myfil_df = df [(df['bpm'] >= 120) & (df['bpm'] <= 130) & (df['danceability_%'] > 50) ]
myfil_df ['bpm'].plot(kind='hist', title= 'Spotify Hit Songs BPM', xlabel= 'BPM Range')
```

```python
myfil_df ['key'].value_counts().plot(kind='pie', title= 'Spotify Hit Songs Key', ylabel='')
```

```python
myfil_df ['mode'].value_counts().plot(kind='pie', title= 'Spotify Hit Songs Mode', ylabel= '')
```

```python
myfil_df ['danceability_%'].plot(kind='hist', title= 'Danceability Percentage for Spotify Hit Songs', xlabel= 'Danceability Percentage')
```

```python
myfil_df ['energy_%'].plot(kind='hist', title= 'Spotify Hit Songs Energy', xlabel= 'Energy')
```

```python
myfil_df ['speechiness_%'].plot(kind='hist', title= 'Speechiness of Spotify Hit Songs', xlabel= 'Speechiness Percentage')
```

```python
myfil_df['acousticness_%'].plot(kind='hist', title= 'Acousticness of Spotify Hit Songs', xlabel= 'Acousticness Percentage')
```
